We create panels, to create rows, columns, and tabs.

Think solar panels, many oriented together.

**The main purpose of Panel's panels is to group things together.**

```python
# initial imports
import pandas as pd
import numpy as np
import panel as pn
from panel.interact import interact
import plotly.express as px

# jupyterlab requires the jupyterlab panel plugin 
# activate using pn.extension
# Without the extension activated, panel objects won't render
pn.extension("plotly")
```

```python
# Create data
housing_transactions = pd.DataFrame(
    {
        "years": np.random.randint(2010, 2019, 100),  # min, max, obs
        "sales": np.random.randint(53, 500, 100),
        "foreclosures": np.random.randint(10, 147, 100),
    }
).sort_values(["years", "sales"])
housing_transactions.describe()
```

```python
# Create scatter plot
scatter_plot = px.scatter(
    housing_transactions,
    x="sales",
    y="foreclosures",
    color="years",
    title="Allegheny Sales/Foreclosures Correlation",
)

# Create bar plot
transactions_by_year = housing_transactions.groupby("years").sum().reset_index()
bar_plot = px.bar(
    transactions_by_year, x="years", y="sales", title="Alleghany Sales by Year"
)

# Create row
row = pn.Row(scatter_plot, bar_plot)
row
```

```python
# Create column using Markdown and row object
column = pn.Column(
    '# Allegheny, PA Real Estate Visualizations',
    '## Sales and Foreclosures',
    row)
column
```

```python
# Create tabs
tabs = pn.Tabs(
    ("Correlations", scatter_plot),
    ("Time Series", bar_plot),
    )
tabs
```
